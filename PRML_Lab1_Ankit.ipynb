```python
##Que1
import math

def is_prime(n):

    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False

    for i in range(3, int(math.sqrt(n)) + 1, 2):
        if n % i == 0:
            return False
    return True

def product_of_primes(numbers):
    prime_product = 1
    found_prime = False

    for num in numbers:
        if is_prime(num):
            prime_product *= num
            found_prime = True

    return prime_product if found_prime else 0


try:
    print("Enter 1000 integers separated by spaces (or fewer for testing):")

    user_list = list(map(int, input().split()))

    if len(user_list) == 0:
        print("The list is empty.")
    else:
        result = product_of_primes(user_list)
        print(f"\nTotal integers processed: {len(user_list)}")
        print(f"Product of all primes: {result}")

except ValueError:
    print("Invalid input. Please ensure you enter only integers.")
```

```python
##Que4

import numpy as np

vectors = [np.random.randn(10,1) for _ in range(5)]

for i in range(5):
  for j in range(i,5):
    inner = np.inner(vectors[i].T, vectors[j].T)
    outer = np.outer(vectors[i], vectors[j])
    print(f"Vectors {i} & {j}: Inner Product = {inner[0][0]:.4f}")


for i in range(5):
  similarities = []
  for j in range(5):
    if i==j: continue

    dot = np.dot(vectors[i].T, vectors[j])[0,0]
    norm_product = np.linalg.norm(vectors[i]) * np.linalg.norm(vectors[j])
    similarity = dot / norm_product
    similarities.append((j, similarity))

  best_sim, best_idx = max(similarities, key=lambda x: x[1])
  print(f"Vectors {i} & {best_sim}: Best Similarity = {best_idx:.4f}")
```

```python
##Que7

import numpy as np
import matplotlib.pyplot as plt
import time

def custom_eig(A, iterations=1000):
  n=A.shape[0]
  V=np.eye(n)
  A_k=A.copy()

  for _ in range(iterations):
    Q,R = np.linalg.qr(A_k)
    A_k = R @ Q
    V = V @ Q

  eigenvalues = np.diag(A_k)
  eigenvectors = V

  return eigenvalues, eigenvectors

np.random.seed(42)
matrix = np.random.rand(5,5)
matrix = (matrix+matrix.T)/2

start=time.time()
custom_vals, custom_vecs = custom_eig(matrix)
custom_time = time.time() - start

start=time.time()
np_vals, np_vecs = np.linalg.eig(matrix)
np_time = time.time() - start

#Results and Comparison
print(f"--- Benchmarking (5x5 Matrix) ---")
print(f"Custom Method Time: {custom_time:.6f}s")
print(f"NumPy Method Time:  {np_time:.6f}s")

#Sort for comparison
custom_vals_sorted = np.sort(custom_vals)
np_vals_sorted = np.sort(np_vals)

print(f"\nSorted Eigenvalues (Custom): \n{custom_vals_sorted}")
print(f"Sorted Eigenvalues (NumPy):  \n{np_vals_sorted}")
print(f"\nMean Absolute Difference: {np.mean(np.abs(custom_vals_sorted - np_vals_sorted)):.2e}")
```

```python
##Que8

import os
from google.colab import drive

# 1. Mount Google Drive
drive.mount('/content/drive')

def display_hierarchy(root_dir, indent="", allowed_exts=None):
    """
    Recursively displays the directory structure, filtering by extensions.
    """
    if allowed_exts is None:
        allowed_exts = {'.jpeg', '.png', '.pdf', '.doc', '.docx', '.ods', '.xls', '.xlsx', '.py', '.mp3', '.mp4'}

    try:
        # Get and sort items to maintain order
        items = sorted(os.listdir(root_dir))
    except PermissionError:
        return
    except FileNotFoundError:
        print("Error: The specified path was not found.")
        return

    for i, item in enumerate(items):
        path = os.path.join(root_dir, item)
        is_last = (i == len(items) - 1)

        # Branching markers
        marker = "└── " if is_last else "├── "

        if os.path.isdir(path):
            # Display folder and recurse
            print(f"{indent}{marker}[Folder] {item}")
            new_indent = indent + ("    " if is_last else "│   ")
            display_hierarchy(path, new_indent, allowed_exts)
        else:
            # Check file extension against allowed list
            _, ext = os.path.splitext(item.lower())
            if ext in allowed_exts:
                print(f"{indent}{marker}{item}")

# --- Set path to your Google Drive ---
path_to_search = "/content/drive/My Drive"

print(f"\nScanning: {path_to_search}")
print("-" * 30)
display_hierarchy(path_to_search)
```

```python
##Que9

import math

class Shape:
    def area(self): pass
    def perimeter(self): pass

class Rectangle(Shape):
    def __init__(self, w, h): self.w, self.h = w, h
    def area(self): return self.w * self.h
    def perimeter(self): return 2 * (self.w + self.h)

class Square(Rectangle):
    def __init__(self, s): super().__init__(s, s)

class Circle(Shape):
    def __init__(self, r): self.r = r
    def area(self): return math.pi * self.r**2
    def perimeter(self): return 2 * math.pi * self.r

class Trapezoid(Shape):
    def __init__(self, a, b, c, d, h): self.a, self.b, self.c, self.d, self.h = a, b, c, d, h
    def area(self): return 0.5 * (self.a + self.b) * self.h
    def perimeter(self): return self.a + self.b + self.c + self.d

class Parallelogram(Shape):
    def __init__(self, b, h, side): self.b, self.h, self.side = b, h, side
    def area(self): return self.b * self.h
    def perimeter(self): return 2 * (self.b + self.side)

def compute(shape):
    print(f"{type(shape).__name__} -> Area: {shape.area():.2f}, Perimeter: {shape.perimeter():.2f}")

# Dynamic Testing
shapes = [
    Circle(7),
    Square(4),
    Rectangle(5, 10),
    Trapezoid(10, 20, 7, 8, 6),
    Parallelogram(12, 8, 10)
]

for s in shapes:
    compute(s)
```

```python
##Que10

import requests
from bs4 import BeautifulSoup
import csv

# Target URL
url = "https://openaccess.thecvf.com/ICCV2025?day=2025-10-23"
base_url = "https://openaccess.thecvf.com"

def scrape_cvf():
    # Fetch the webpage
    response = requests.get(url)
    if response.status_code != 200:
        print("Failed to retrieve the webpage.")
        return

    soup = BeautifulSoup(response.text, 'html.parser')

    # Extract Titles, Authors, and PDF Links

    titles = [t.text.strip() for t in soup.find_all('dt', class_='ptitle')]

    dd_tags = soup.find_all('dd')

    authors_list = []
    pdf_links = []

    for i in range(0, len(dd_tags), 2):
        # Extract Authors (cleaning up extra spaces/newlines)
        author_text = dd_tags[i].text.replace('\n', '').strip()
        authors_list.append(author_text)

        # Extract PDF Link
        pdf_tag = dd_tags[i+1].find('a', string='pdf')
        if pdf_tag:
            pdf_links.append(base_url + pdf_tag['href'])
        else:
            pdf_links.append("N/A")

    # Save to CSV
    filename = "ICCV2025_Papers.csv"
    with open(filename, 'w', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow(["Paper Title", "Authors", "PDF Link"])

        # Zip the lists and write rows
        for data in zip(titles, authors_list, pdf_links):
            writer.writerow(data)

    print(f"Successfully saved {len(titles)} papers to {filename}")

if __name__ == "__main__":
    scrape_cvf()
```
